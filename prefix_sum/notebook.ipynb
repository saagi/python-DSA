2. Subarray Sum Equals K

Find the number of continuous subarrays whose sum equals k.

Example:
Input: nums = [1, 1, 1], k = 2
Output: 2
(These subarrays sum to 2: [1,1] at positions 0–1 and 1–2)

```python
## Brute Force
def num_subarrays(nums,k):
    count=0
    for start in range(len(nums)):
        for end in range(start,len(nums)):
            if sum(nums[start:end+1])==k:
                count+=1
    return count

num_subarrays([1,1,1,1,2,1],5)


```

```python
def subarray_sum(nums, k):
    prefix_count = {0: 1}
    prefix_sum = 0
    count = 0

    for num in nums:
        prefix_sum += num
        
        # Check if there exists a previous prefix sum
        # such that prefix_sum - previous = k
        if prefix_sum - k in prefix_count:
            count += prefix_count[prefix_sum - k]

        # Update frequency
        prefix_count[prefix_sum] = prefix_count.get(prefix_sum, 0) + 1
        print(f"prefix_sum = {prefix_sum}, count = {count}, prefix_count = {[prefix_count]}")

    return count
subarray_sum([1,1,1,2,3],5)
```

```python
def prefix_sum(nums,k):
    prefix_s = [0]*len(nums)
    prefix_s[0] = nums[0]
    for i in range(1,len(nums)):
        prefix_s[i]+=prefix_s[i-1]+nums[i]
    return prefix_s
```

Given a 0-indexed string word and a character ch, reverse the segment of word that starts at index 0 and ends at the index of the first occurrence of ch (inclusive). If the character ch does not exist in word, do nothing.

For example, if word = "abcdefd" and ch = "d", then you should reverse the segment that starts at 0 and ends at 3 (inclusive). The resulting string will be "dcbaefd".
Return the resulting string.

 

Example 1:

Input: word = "abcdefd", ch = "d"
Output: "dcbaefd"
Explanation: The first occurrence of "d" is at index 3. 
Reverse the part of word from 0 to 3 (inclusive), the resulting string is "dcbaefd".
Example 2:

Input: word = "xyxzxe", ch = "z"
Output: "zxyxxe"
Explanation: The first and only occurrence of "z" is at index 3.
Reverse the part of word from 0 to 3 (inclusive), the resulting string is "zxyxxe".

```python
## Solution-1

class Solution:
    def reversePrefix(self, word: str, ch: str) -> str:
        found=False
        for i in range(len(word)):
            if word[i]==ch:
                found=True
                break
        if found:
            return word[i::-1]+word[i+1:]
        return word
```

```python
## Solution-2
class Solution:
    def reversePrefix(self, word: str, ch: str) -> str:
        for i in range(len(word)):
            if ch == word[i]:
                return word[i::-1]+word[i+1:]
        return word
```

```python
def longest_subarray_sum_k(arr, k):
    """
    Returns the length of the longest subarray with sum equal to k.

    Approach:
    - Use prefix sum + hashmap
    - Store first occurrence of each prefix sum
    - For each index, check if (prefix_sum - k) was seen before

    Time Complexity: O(n)
    Space Complexity: O(n)
    """

    prefix_sum = 0
    max_len = 0
    prefix_map = {}  # stores: prefix_sum -> first index

    for i, num in enumerate(arr):
        prefix_sum += num

        # Case 1: subarray from index 0 to i
        if prefix_sum == k:
            max_len = i + 1

        # Case 2: subarray from some previous index + 1 to i
        if (prefix_sum - k) in prefix_map:
            length = i - prefix_map[prefix_sum - k]
            max_len = max(max_len, length)

        # Store first occurrence only (to maximize length)
        if prefix_sum not in prefix_map:
            prefix_map[prefix_sum] = i

    return max_len
```

```python
def subarrays_divisible_by_k(arr, k):
    """
    Returns the count of subarrays whose sum is divisible by k.

    Approach:
    - Use prefix sum
    - Track remainders (prefix_sum % k)
    - If same remainder appears again, subarray between them is divisible by k

    Time Complexity: O(n)
    Space Complexity: O(n)
    """

    prefix_sum = 0
    count = 0
    remainder_map = {0: 1}  # remainder → frequency

    for num in arr:
        prefix_sum += num

        # Compute remainder
        remainder = prefix_sum % k

        # Handle negative remainder (safe for all languages)
        if remainder < 0:
            remainder += k

        # If remainder seen before, add its frequency
        if remainder in remainder_map:
            count += remainder_map[remainder]

        # Update frequency of remainder
        remainder_map[remainder] = remainder_map.get(remainder, 0) + 1

    return count
```
